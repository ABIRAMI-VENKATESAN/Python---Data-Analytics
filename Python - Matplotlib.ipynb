# LINE PLOT DEMO:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,10,1)
y = 2*x + 5
plt.plot(x,y,color ='green',marker='*')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,10,1)
y = 2*x + 5
fig = plt.figure(figsize = (10,5))
plt.plot(x, y, linewidth = 2.0, linestyle='-.', color = 'm',alpha = 0.7,marker='D')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,10,1)
y1 = 2*x + 5
y2 = 3*x + 10
plt.subplot(1,2,1)
plt.plot(x,y1,color='r')
plt.title('Graph1')
plt.subplot(1,2,2)
plt.plot(x,y2,color = 'b')
plt.title('Grape2')
plt.show()
```

# HOW TO CREATE A SCATTER PLOT:

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
a = np.arange(10,90,10)
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]
plt.scatter(a,b,color='b')
plt.scatter(a,x,color = 'r')
plt.show()
```

# HISTOGRAM:

```python
import matplotlib.pyplot as plt
%matplotlib inline
age = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
plt.hist(age,bins=[0,20,40,60,80,100])
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
age = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
plt.hist(age,bins = [0,20,40,60,80,100],edgecolor = '#000000',color = '#FF2331')
plt.title("Histogram Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequency of Values")
plt.grid(True)
plt.savefig("Histogram Plot.pdf")
plt.show()
```

# PIE CHART:

```python
import matplotlib.pyplot as plt
%matplotlib inline
animal = ['Dog','Cat','Wolf','Lion']
count = [50,45,60,80]
plt.pie(count,labels = animal)
plt.title("Pie Chart Demo" )
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
animal = ['Dog','Cat','Wolf','Lion']
count = [50,45,60,80]
plt.pie(count,labels = animal,autopct = "%1.1f%%",shadow = True, startangle = 90,explode = (0,0.1,0,0))
plt.title("Pie Chart Demo" )
plt.show()
```
